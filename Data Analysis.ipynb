### Importing required libraries

```python
import os
import spacy
from collections import Counter
import nltk
from nltk import ngrams
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
```

### Defining the list of emotions and stopwords

```python
# Define the list of emotions
emotions = ['Angry', 'Happy', 'Neutral', 'Sad']

# Define a list of Urdu stopwords
# Note: This is just an example. You should replace this with a comprehensive list of Urdu stopwords.
urdu_stopwords = ["اس" ,"اور", "کی", "میں", "کے", "سے", "ہے", "کو", "پر", "ہی", "ہو", "یہ", "تھا", "کر", "نے", "بھی", "تو", "جو", "ہم", "مگر", "کچھ", "ہوا", "کہ", "اب", "ہے۔", "جب", "ہوتا", "ہوتی", "ہوتے", "ہونے", "و", "وہ", "یا", "تھی", "تھے", "تھیں", "نہیں" , "ہیں", "آپ" , "کا", "ہے"]
```

## Part 2 EDA

### a) Top 10 most frequent words in the transcribed text for each emotion

```python
# Iterate over each emotion
for emotion in emotions:
    # Initialize a Counter for word frequencies
    word_freq = Counter()

    # Get the list of text files for this emotion
    text_files = os.listdir(f'./Speech To Text/Merged_{emotion}_Text')

    for text_file in text_files:
        with open(f'./Speech To Text/Merged_{emotion}_Text/{text_file}', 'r', encoding='utf-8') as file:
            text = file.read()
            # Tokenize the text into individual words
            words = nltk.word_tokenize(text)
            # Remove stopwords from the list of words
            words = [word for word in words if word not in urdu_stopwords]
            # Update the word frequency Counter
            word_freq.update(words)

    # Get the 10 most common words
    common_words = word_freq.most_common(10)

    # Split the words and their frequencies into two lists for plotting
    words, frequencies = zip(*common_words)
    
    # Reshape the words for displaying
    reshaped_words = [get_display(arabic_reshaper.reshape(word)) for word in words]

    # Plot the frequencies
    plt.figure(figsize=(10,5))
    plt.bar(reshaped_words, frequencies)
    plt.title(get_display(arabic_reshaper.reshape(f'Top 10 Most Frequent Words for {emotion}')))
    plt.show()
```

### b) Word Cloud Analysis

```python
# Iterate over each emotion
for emotion in emotions:
    # Get the list of text files for this emotion
    text_files = os.listdir(f'./Speech To Text/Merged_{emotion}_Text')
    all_text = ""
    for text_file in text_files:
        with open(f'./Speech To Text/Merged_{emotion}_Text/{text_file}', 'r', encoding='utf-8') as file:
            text = file.read()
            all_text += text

    # Tokenize the text into individual words
    words = nltk.word_tokenize(all_text)
    
    # Join the list of words back into a single string with spaces in between
    all_text = " ".join(words)

    # Generate wordcloud
    wordcloud = WordCloud(width = 600, height = 600, 
                    background_color ='white', font_path='NotoNaskhArabic-Regular.ttf', 
                    stopwords = None,
                    min_font_size = 10).generate(get_display(arabic_reshaper.reshape(all_text)))

    # Plot the WordCloud image                        
    plt.figure(figsize = (6, 6), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0)
    plt.title(f'Word Cloud for {emotion}')
    plt.show()
```

### c) Average length of the transcribed text for each emotion

```python
# Initialize a dictionary to hold the average lengths for each emotion
avg_lengths = {}

# Iterate over each emotion
for emotion in emotions:
    # Initialize a list to hold the lengths of the transcribed texts
    lengths = []

    # Get the list of text files for this emotion
    text_files = os.listdir(f'./Speech To Text/Merged_{emotion}_Text')

    for text_file in text_files:
        with open(f'./Speech To Text/Merged_{emotion}_Text/{text_file}', 'r', encoding='utf-8') as file:
            text = file.read()
            # Add the length of the transcribed text to the list
            lengths.append(len(text))

    # Calculate the average length for this emotion and add it to the dictionary
    avg_lengths[emotion] = sum(lengths) / len(lengths)

# Split the emotions and their average lengths into two lists for plotting
emotions, avg_lengths = zip(*avg_lengths.items())

# Plot the average lengths
plt.figure(figsize=(10,5))
plt.bar(emotions, avg_lengths)
plt.title('Average Length of Transcribed Text for Each Emotion')
plt.show()
```

### d) Vader Sentiment Analysis

```python
# Initialize the sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Initialize a dictionary to hold the average sentiment scores for each emotion
avg_scores = {}

# Iterate over each emotion
for emotion in emotions:
    # Initialize a list to hold the sentiment scores of the transcribed texts
    scores = []

    # Get the list of text files for this emotion
    text_files = os.listdir(f'./Speech To Text/Merged_{emotion}_Text')

    for text_file in text_files:
        with open(f'./Speech To Text/Merged_{emotion}_Text/{text_file}', 'r', encoding='utf-8') as file:
            text = file.read()
            # Calculate the sentiment polarity of the text
            polarity_scores = sia.polarity_scores(text)
            # Add the compound sentiment score to the list
            scores.append(polarity_scores['compound'])

    # Calculate the average sentiment score for this emotion and add it to the dictionary
    avg_scores[emotion] = sum(scores) / len(scores)

# Split the emotions and their average sentiment scores into two lists for plotting
emotions, avg_scores = zip(*avg_scores.items())

print(avg_scores)

# Plot the average sentiment scores
plt.figure(figsize=(10,5))
plt.plot(emotions, avg_scores, marker='o')
plt.title('Average Sentiment Polarity of Transcribed Text for Each Emotion')
plt.ylabel('Average Sentiment Polarity')
plt.grid(True)
plt.show()
```

### e) Top 10 most frequent bigrams and trigrams

```python
from arabic_reshaper import reshape

# Iterate over each emotion
for emotion in emotions:
    # Get the list of text files for this emotion
    text_files = os.listdir(f'./Speech To Text/Merged_{emotion}_Text')
    all_text = ""
    for text_file in text_files:
        with open(f'./Speech To Text/Merged_{emotion}_Text/{text_file}', 'r', encoding='utf-8') as file:
            text = file.read()
            all_text += text

    # Tokenize the text into individual words
    words = nltk.word_tokenize(all_text)
    
    # Generate bigrams and trigrams
    bigrams = ngrams(words, 2)
    trigrams = ngrams(words, 3)
    
    # Calculate the frequencies
    bigram_freq = Counter(bigrams)
    trigram_freq = Counter(trigrams)
    
    # Get the 10 most common bigrams and trigrams
    common_bigrams = bigram_freq.most_common(10)
    common_trigrams = trigram_freq.most_common(10)
    
    # Split the bigrams/trigrams and their frequencies into two lists for plotting
    bigrams, bigram_frequencies = zip(*common_bigrams)
    trigrams, trigram_frequencies = zip(*common_trigrams)
    
    # Reshape the bigrams and trigrams for displaying
    reshaped_bigrams = [get_display(reshape(' '.join(bigram))) for bigram in bigrams]
    reshaped_trigrams = [get_display(reshape(' '.join(trigram))) for trigram in trigrams]
    
    # Plot the frequencies
    plt.figure(figsize=(10,5))
    plt.bar(reshaped_bigrams, bigram_frequencies)
    plt.title(get_display(reshape(f'Top 10 Most Frequent Bigrams for {emotion}')))
    plt.show()

    plt.figure(figsize=(10,5))
    plt.bar(reshaped_trigrams, trigram_frequencies)
    plt.title(get_display(reshape(f'Top 10 Most Frequent Trigrams for {emotion}')))
    plt.show()
```

## Part 3 NLP Downstream Task

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# Initialize the vectorizer and classifier
vectorizer = TfidfVectorizer()
classifier = LinearSVC()

# Lists to store the texts and labels
texts = []
labels = []

# Iterate over each emotion
for emotion in emotions:
    # Get the list of text files for this emotion
    text_files = os.listdir(f'./Speech To Text/Merged_{emotion}_Text')

    for text_file in text_files:
        with open(f'./Speech To Text/Merged_{emotion}_Text/{text_file}', 'r', encoding='utf-8') as file:
            text = file.read()
            # Append the text and label to the respective lists
            texts.append(text)
            labels.append(emotion)

# Convert the text data into numeric data
X = vectorizer.fit_transform(texts)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Train the classifier
classifier.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = classifier.predict(X_test)

# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```
